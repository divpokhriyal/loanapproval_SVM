```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset 

loan_dataset = pd.read_csv('loan_approval.csv')
```

```python
loan_dataset
```

```python
#statistical measures of the dataset

loan_dataset.describe()
```

```python
#number of missing values in each column

loan_dataset.isnull().sum()
```

```python
#dropping the missing values

loan_dataset = loan_dataset.dropna()
```

```python
loan_dataset
```

```python
#label encoding using Pandas - replacing Y with 1 and N with no

loan_dataset.replace({"Loan_Status":{'N':0, 'Y':1}}, inplace = True)
```

```python
loan_dataset
```

```python
#Dependent column values
loan_dataset['Dependents'].value_counts()
```

```python
#Replacing the value 3+ with 4 for better processing

loan_dataset.replace({"Dependents":{'3+':4}}, inplace = True)
```

```python
loan_dataset
```

```python
loan_dataset['Dependents'].value_counts()
```

#### Visualizing the data 

```python
#education and loan status

sns.countplot(x='Education', hue = 'Loan_Status', palette=['#432371',"#FAAE7B"], data = loan_dataset)
```

```python
#marital status and loan status 
sns.set_theme(style="darkgrid")
sns.countplot(x='Married', hue = 'Loan_Status', color = 'darkred', data = loan_dataset)
```

```python
#Self_Employed and loan status
sns.set_theme(style="darkgrid")
sns.countplot(x='Self_Employed', hue = 'Loan_Status', palette="Set1", data = loan_dataset)
```

```python
#Converting categorical columns into numerical values

loan_dataset.replace({'Married':{'No':0, 'Yes':1}, 'Gender':{'Male':1, 'Female':0}, 'Self_Employed':{'No':0, 'Yes':1}, 'Property_Area':{'Rural':0, 'Semiurban':1, 'Urban':2}, 'Education':{'Graduate':1, 'Not Graduate':0}}, inplace = True)
```

```python
loan_dataset
```

```python
#Seperating the data and label

X = loan_dataset.drop(columns=['Loan_ID', 'Loan_Status'], axis=1)
Y = loan_dataset['Loan_Status']
```

```python
X
```

```python
Y
```

```python
#splitting data into training and testing data -- X- data, Y- label

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#training the model - Support Vector Machine model

classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train, Y_train)
```

```python
#Model evaluation - accuracy score

X_train_prediction = classifier.predict(X_train)

training_data_accuracy = accuracy_score(X_train_prediction, Y_train)

print('Accuracy of training data = ', training_data_accuracy)
```

```python
#Accuracy of testing data

X_test_prediction = classifier.predict(X_test)

testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)

print('Accuracy of testing data = ', testing_data_accuracy)
```
